### Web Scraping

Scraping the required Wikipedia page: 

The Pandas library in Python includes a web scraper that pulls HTML table data into a dataframe in a single step. Simply insert the URL into the read_html() method and assign the resulting object to a variable.

```python
data = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
```

We saved the scraped data into the variable 'data'. Letâ€™s find out how many tables it contains.

```python
print(len(data)) # To check the number of tables in the 
```

There are 3 tables in total. Let us check the first one.

```python
data = data[0] #To check first table
data
```

```python
#Assigning the first row to be the table header
df = data.rename(columns=data.iloc[0])
df
```

```python
#Drop the first row

df = df.drop([0], axis=0)  # '[0]'' denotes first row. 'axis = 0' denotes 'rows'
df
```

```python
# How many rows have Borough equal to 'Not assigned'?
df[df['Borough']=='Not assigned'].count()
```

```python
#Ignore cells with a borough that is 'Not assigned'.

df2 = df[df.Borough != 'Not assigned'] 
df2
```

```python
# Reset index
df2.reset_index(drop=True, inplace=True)
df2
```
